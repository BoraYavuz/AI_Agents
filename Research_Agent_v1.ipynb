```python
from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage

# Test OpenAI connection
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
response = llm.invoke([HumanMessage(content="Hello! Can you help me with research?")])

print("🤖 AI Response:", response.content)
print("✅ OpenAI connection working!")
```

```python
import arxiv
from datetime import datetime

def search_papers(topic, max_results=100):
    """Search arXiv for papers on a topic"""
    # Use arxiv.Client() instead of arxiv.Search() to avoid DeprecationWarning
    client = arxiv.Client()

    search = arxiv.Search(
        query=topic,
        max_results=max_results,
        # sort_by=arxiv.SortCriterion.SubmittedDate
        sort_by=arxiv.SortCriterion.Relevance,
        sort_order=arxiv.SortOrder.Descending
    )

    papers = []
    # Use client.results() instead of search.results()
    for paper in client.results(search):
        papers.append({
            'title': paper.title,
            'authors': [author.name for author in paper.authors][:3],  # First 3 authors
            'published': paper.published.strftime('%Y-%m-%d'),
            'summary': paper.summary[:1200] + "...",  # First 1200 chars
            'url': paper.entry_id
        })

    return papers

# Test the paper search
topic = "AI agent applications review"  # Change this to your interest
papers = search_papers(topic, 50)

print(f"📚 Found {len(papers)} papers on '{topic}':\n")
for i, paper in enumerate(papers, 1):
    print(f"{i}. **{paper['title']}**")
    print(f"   Authors: {', '.join(paper['authors'])}")
    print(f"   Published: {paper['published']}")
    print(f"   Summary: {paper['summary']}")
    print(f"   URL: {paper['url']}\n")
```

```python
def research_assistant(query):
    """Simple research assistant"""
    print(f"🔍 Researching: {query}")
    print("=" * 50)

    # Step 1: Search for papers
    papers = search_papers(query, 5)

    # Step 2: Create summary with AI
    paper_info = "\n".join([f"- {p['title']} ({p['published']}): {p['summary']}" for p in papers])

    prompt = f"""Based on these recent papers about {query}:

{paper_info}

Please provide:
1. A brief overview of the current research landscape
2. Key findings and themes
3. Potential research gaps or future directions

Keep it concise but informative."""

    # Step 3: Get AI analysis
    response = llm.invoke([HumanMessage(content=prompt)])

    print("📊 RESEARCH ANALYSIS:")
    print(response.content)
    print("\n" + "=" * 50)
    print(f"📚 Based on {len(papers)} recent papers")

    return papers, response.content

# Test it out
research_topic = input("Enter a research topic: ")
papers, analysis = research_assistant(research_topic)
```

```python
import pandas as pd
import plotly.express as px
from collections import Counter

def analyze_research_trends(topic, max_papers=20):
    """Analyze research trends for a topic"""
    papers = search_papers(topic, max_papers)

    if not papers:
        return "No papers found!"

    # Convert to DataFrame for analysis
    df = pd.DataFrame(papers)
    df['year'] = pd.to_datetime(df['published']).dt.year

    # Publication trends
    yearly_counts = df.groupby('year').size().reset_index(name='papers')

    # Create visualization
    fig = px.bar(yearly_counts, x='year', y='papers',
                 title=f'Publication Trends: {topic}')
    fig.show()

    # Author analysis
    all_authors = []
    for authors in df['authors']:
        all_authors.extend(authors)

    author_counts = Counter(all_authors).most_common(5)

    print(f"📈 TREND ANALYSIS for '{topic}':")
    print(f"Total papers analyzed: {len(papers)}")
    print(f"Year range: {yearly_counts['year'].min()} - {yearly_counts['year'].max()}")
    print(f"Most productive year: {yearly_counts.loc[yearly_counts['papers'].idxmax(), 'year']}")
    print(f"\n👥 Most frequent authors:")
    for author, count in author_counts:
        print(f"  - {author}: {count} papers")

    return df

# Test trend analysis
topic = input("Enter topic for trend analysis: ")
df = analyze_research_trends(topic, 100)
```
